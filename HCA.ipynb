```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage 
import seaborn as sns
from seaborn import clustermap
```

```python
# Load both datasets
D_data = pd.read_csv("dengue_data.csv", index_col=0)
D_metadata = pd.read_csv("dengue_metadata.csv",index_col=0)
```

```python
# Perform HCA
data_transposed = D_data.transpose()
linked = linkage(data_transposed, method='complete', metric='euclidean') 

# Sample-centric dendrogram
plt.figure(figsize=(15, 8))
dendrogram(linked, labels=data_transposed.index, orientation='top') 
plt.title("Sample-Centric Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

# Perform HCA on the first 100 genes
first100genes = D_data.iloc[:100, :]
HCA100genes = linkage(first100genes, method='complete', metric='euclidean') 

# Gene-centric dendrogram
plt.figure(figsize=(15, 8))
dendrogram(HCA100genes, labels=first100genes.index, orientation='top')
plt.title("Gene-Centric Dendrogram (First 100 Genes)")
plt.xlabel("Genes")
plt.ylabel("Distance")
plt.show()
```

```python
# Plot heatmap
heatmap = clustermap(first100genes, method='complete', metric='euclidean',cmap='viridis') 
plt.show()
```
